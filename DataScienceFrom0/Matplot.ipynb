# Matplop

```python
from matplotlib import pyplot as plt
```

```python
years = range(1950,2011,10)
gdp = [(x / 0.15) * (i**1.5) for i,x  in enumerate(years)]
```

```python
plt.plot(years, gdp,color='red', marker='o', linestyle='solid')
plt.title("GDP Nominal")
plt.ylabel("Bilhoes de $")
plt.show()
```

## Bar Charts

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]
```

```python
xs = [i + 0.1 for i , _ in enumerate(movies)]
```

```python
plt.bar(xs, num_oscars)
plt.title("Oscars")
plt.xticks([i + 0.5 for i,_ in enumerate(movies)], movies) #nomeia os eixos com os nomes dos filmes
plt.show()
```

```python
from collections import Counter
```

```python
grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]
decile = lambda grade: grade // 10 * 10 
histrogram = Counter(decile(grade) for grade in grades)
```

```python
plt.bar([x-4 for x in histrogram.keys()],
       histrogram.values(),
       8)
plt.axis([-5,102,0,5])
plt.xticks(range(10,121,10), color='white')
plt.yticks(color='white')
plt.xlabel('Decil', color='white')
plt.ylabel('Students', color='white')
plt.title('Grades Distribution', color='white')
plt.show()
```

## Line Charts

```python
variance     = [1,2,4,8,16,32,64,128,256]
bias_squared = [256,128,64,32,16,8,4,2,1]
total_error  = [x + y for x, y in zip(variance, bias_squared)]
xs = [x for x , _ in enumerate(variance)]
```

### multiplas chamadas de plot para mostrar multiplas series

```python
solid_green = 'g-'
dashed_red ='r-.'
pointed_blue = 'b:'
plt.plot(xs, variance, solid_green, label='variance')
plt.plot(xs, bias_squared, dashed_red, label='bias ^ 2')
plt.plot(xs, total_error, pointed_blue, label='total error')
```

```python
plt.legend(loc=9) #loc=9 -> top center
plt.xlabel('model complexity', color='white')
plt.title('Commitment between polarization and variance', color='white')
plt.xticks(color='white')
plt.yticks(color='white')
plt.show()
```

## Scatter Charts

```python
friends = [ 70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
```

```python
plt.scatter(friends, minutes)
```

```python
#nomeia cada posicao
for label, friend_count, minute_count in zip(labels,friends, minutes):
    plt.annotate(label,
                xy=(friend_count, minute_count),
                xytext=(5,-5),
                textcoords='offset points')
```

```python
plt.title('Daily Minutes vc # of Friends', color='white')
plt.xlabel('# of Friends', color='white')
plt.ylabel('daily minutes spent on site',color='white')
plt.xticks(color='white')
plt.yticks(color='white')
plt.show()
```
