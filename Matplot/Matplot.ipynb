```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

### Functional Method

```python
plt.plot(x,y)
plt.xlabel('X', color='white')
plt.ylabel('Y', color='white')
plt.title('Title', color='white')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(x,y,'y')
```

## OO Method

```python
fig = plt.figure() ## instance of a blank canvas
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y, 'r')
axes.set_xlabel('X', color='white')
axes.set_ylabel('Y', color='white')
axes.set_axis_bgcolor('black')
```

```python
fig = plt.figure() ## instance of a blank canvas
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y, 'r')
axes1.set_xlabel('X', color='white')
axes1.set_ylabel('Y', color='white')
axes1.set_axis_bgcolor('gray')
axes1.set_title('OUTER', color='white')

axes2 = fig.add_axes([0.4,0.3,0.4,0.3])
axes2.plot(y,x, 'white')
axes2.set_xlabel('X', color='white')
axes2.set_ylabel('Y', color='white')
axes2.set_axis_bgcolor('blue')
axes2.set_title('INNER', color='yellow')
```

## Subplots

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
for i, a in  enumerate(axes):
    plot_color = 'red'
    if i % 2 == 0:
        plot_color = 'green'
    a.plot(x,y, plot_color)
    a.set_xlabel('X', color='white')
    a.set_ylabel('Y', color='white')
    bg_color = 'gray'
    if i % 2 == 0:
        bg_color = 'yellow'
    a.set_axis_bgcolor(bg_color)
    plot_title = 'Plot ' + str((i+1))
    a.set_title(plot_title, color='white')
    
plt.tight_layout()
```

## Figure Size and DPI

```python
import numpy as np

fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(11,8))
for i, a in  enumerate(axes):
    list = range(1,4)
    for n in list:
        r = np.random.randint(1000)
        exp = n+(i*r)
        if i % 2 == 0:
            exp = n+r
        a.plot(x,y*exp, label= 'X*'+str(exp))
        
    a.set_xlabel('X', color='white')
    a.set_ylabel('Y', color='white')
    bg_color = 'black'
    if i % 2 == 0:
        bg_color = 'yellow'
    a.set_axis_bgcolor(bg_color)
    plot_title = 'Plot ' + str((i+1))
    a.set_title(plot_title, color='white')
    a.legend(loc=0)
    
plt.tight_layout()
```

```python
fig.savefig('my_picture.jpg', dpi=200)
```

## Look and Fell

```python
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(11,8))
for i, a in  enumerate(axes):
    list = range(1,4)
    for n in list:
        r = np.random.randint(1000)
        exp = n-(i*r)
        if i % 2 == 0:
            exp = n+r
        a.plot(x,y*exp, label= 'X*'+str(exp), 
               linewidth=n*1.5, #look and feel opts
               alpha=0.7, 
               linestyle='-.', 
               marker='o',
              markersize=10,
              markerfacecolor='black',
              markeredgewidth=3,
              markeredgecolor='yellow')
        
    a.set_xlabel('X', color='white')
    a.set_ylabel('Y', color='white')
    bg_color = 'black'
    if i % 2 == 0:
        bg_color = 'white'
    a.set_axis_bgcolor(bg_color)
    plot_title = 'Plot ' + str((i+1))
    a.set_title(plot_title, color='white')
    a.legend(loc=0)
    
plt.tight_layout()
```
